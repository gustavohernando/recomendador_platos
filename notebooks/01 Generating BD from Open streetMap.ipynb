Ataca a la API de OpenStreetmap con el siguiente código de ejemplo

```python
def json_to_df(data):
    elements = data['elements']
    places = {'id_OSM':[],'name':[],
              'tipo':[],'cuisine':[],'vegan':[],'vegetarian':[],
              'lat':[],'lon':[],'address': [], 'postcode': [],
              'phone': [],'website': [],'facebook': [],
              'opening_hours': [],'image': []
             }
    # OSM  Open street map
    
    for i in elements:
        vegan= i.get('tags', None).get('diet:vegan',"no")
        vegetarian= i.get('tags', None).get('diet:vegetarian',"no")
        
        if (vegan != "no") or (vegetarian != "no"):
            id_OSM = i.get('id', None)
            tipo = i.get('tags', None).get('amenity',None)
            latitude = i.get('lat', None)
            longitude = i.get('lon', None)
            cuisine = i.get('tags', None).get('cuisine',None)
            name = i.get('tags', {}).get('name', "NO NAME")
            street = i.get('tags', {}).get('addr:street', "NO STREET")
            number = i.get('tags', {}).get('addr:housenumber', 9999)
            postcode = i.get('tags', {}).get('addr:postcode', None)
            phone = i.get('tags', None).get('contact:phone',None)
            facebook = i.get('tags', None).get('contact:facebook',None)
            website = i.get('tags', None).get('website',None)
            opening_hours = i.get('tags', None).get('opening_hours:',None)
            image = i.get('tags', None).get('image',None)

            places['id_OSM'].append(id_OSM)
            places['cuisine'].append(cuisine)
            places['tipo'].append(tipo)
            places['lat'].append(latitude)
            places['lon'].append(longitude)
            places['name'].append(name)
            places['address'].append(street + ' ' + str(number))
            places['postcode'].append(postcode)
            places['phone'].append(phone)
            places['facebook'].append(facebook)
            places['website'].append(website)
            places['vegan'].append(vegan)
            places['vegetarian'].append(vegetarian)
            places['opening_hours'].append(opening_hours)
            places['image'].append(image)
        
    return pd.DataFrame(places)
```

```python
overpass_url = "http://overpass-api.de/api/interpreter"
overpass_query = overpass_query = """
[out:json];
node["amenity"~"cafe|restaurant|fast_food"]
  (40.396013, -3.746999, 40.480011, -3.625910);
out;
"""
response = requests.get(overpass_url, 
                        params={'data': overpass_query})
print(response)
data_2 = response.json()
```

```python
bdrest_json = response.json()
bdrest_df = json_to_df(data_2)
bdrest_df.head()
```
